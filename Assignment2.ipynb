1. Download the dataset: Dataset
2. Load the dataset.

```python
#Read the dataset into a pandas dataframe
df=pd.read_csv("/content/House Price India.csv")
df.head()
```

3. Perform the Below Visualizations.
**● Univariate Analysis**

```python
sns.distplot(df['living area'])
```

```python
sns.displot(df['lot area'])
```

```python
plt.pie(df['number of bedrooms'].value_counts(),labels=df['number of bedrooms'].value_counts().index,autopct='%1.1f%%',shadow=True)
plt.title('No of bedrooms Pie Chart')
plt.show()
```

**● Bi - Variate Analysis**

```python
sns.lineplot(x=df['number of bedrooms'],y=df['Price'])
```

```python
sns.scatterplot(x=df['living area'],y=df['Price'])
```

**● Multivariate Analysis**

```python
sns.pairplot(df)
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(df.corr(),annot=True)
```
